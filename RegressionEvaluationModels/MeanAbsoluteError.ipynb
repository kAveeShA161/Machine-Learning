```python
import numpy as np 
import pandas as pd
```

```python
df = pd.DataFrame({
    'hours':[8,11,9,5,7.5,9.5,10,7,9,9.5,8,10.5,9,6.5,9],
    'score':[56,70,51,24,30,66,48,36,42,61,39,87,73,48,46]
})
```

```python
y = np.array(df.score.values)
```

```python
y_pred = 8 * df.hours.values - 15
```

```python
from sklearn.metrics import mean_absolute_error
meann = mean_absolute_error(y, y_pred)
print(f"Mean Absolute Error: {meann}")
```

```python
from sklearn.metrics import root_mean_squared_error
rmse = root_mean_squared_error(y, y_pred)
print(f"Root Mean Squared Error: {rmse}")
```

```python
from sklearn.metrics import r2_score
r2s = r2_score(y, y_pred)
print(f"R2_Score: {r2s}")
```
